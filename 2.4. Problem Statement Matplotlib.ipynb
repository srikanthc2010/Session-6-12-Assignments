### 1. Create a pie chart presenting the male/female proportion

```python
import pandas as pd
import matplotlib.pyplot as plt

dfTitanic = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
maleFemaleCount = dfTitanic.groupby('sex')['sex'].count()
labs = ['Female','Male']

plt.pie(maleFemaleCount,labels=labs,explode= (0,0.1),shadow= True,autopct='%1.1f%%',startangle = 160)
plt.show()
```

### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
femaleDataF = dfTitanic[dfTitanic.sex == 'female']
xAge = femaleDataF['age']
yFare = femaleDataF['fare']
colors = ("blue")
groups = ('Female')
plt.xlabel('Age of Pasengers')
plt.ylabel('Fare paid by passegers')
plt.scatter(xAge,yFare,c=colors, label = groups)

maleDataF = dfTitanic[dfTitanic.sex == 'male']
xAge = maleDataF['age']
yFare = maleDataF['fare']
colors = ("red")
groups = ('Male')
plt.scatter(xAge,yFare,c=colors,label = groups)
plt.title('Tatanic fare paid and age of passengers')
plt.legend()
plt.show()
```
