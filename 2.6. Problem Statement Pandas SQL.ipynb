# 5. What is the average, minimum and maximum age group for people working in different sectors

```python
query = "SELECT workclass, avg(age) as Avg_Age,min(age) as Min_Age,max(age) as Max_age FROM adultDf GROUP BY workclass;"
pysqldf(query)
```

# 6. Calculate age distribution by country

```python
query = """SELECT [native-country], Round(Round((count(age)*100),5)/Round((select COUNT(age) 
from adultDf),5),5) as "Age Distribution"  FROM adultDf group by [native-country] order by count(age) desc;"""
pysqldf(query)
```

# 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

# 1. Create an sqlalchemy engine using a sample from the data set

```python
import sqlalchemy as db
import pandas as pd
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", delimiter=',')
# Add column headers
df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'label']

```

```python
# 1. Create sqlalchemy Engine
engine = db.create_engine('sqlite:///mydb1', echo =False)

# Create table using dataframe and engine-- load data
df.to_sql('Adult', engine, if_exists='replace', chunksize=1000)
```

# 2. Write two basic update queries

```python
# Print to see the existing values
result = engine.execute("SELECT fnlwgt, education,age FROM Adult WHERE age = 39").fetchall()
print("{l} records found".format(l=len(result)))
print(result)
print("\n")
```

```python
#two basic update quries
#update adult table recordes of ecduation values to Technical where age is 39
engine.execute("UPDATE Adult set education = 'Technical' WHERE age = 39")

#update adult table of fnlwgt columns values to 0 where age  = 39
engine.execute("UPDATE Adult set fnlwgt = 0 WHERE age = 39")
```

```python
#Checking Reslut below: Updated education and fnlwgt column values.
result = engine.execute("SELECT fnlwgt, education,age FROM Adult WHERE age = 39").fetchall()
print("{l} records found".format(l=len(result)))
print(result)
print("\n")
```

# 5. Write two function queries

```python

# Define functions
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

# Create a new table with Column named "Value"
engine.execute('CREATE TABLE IF NOT EXISTS Fib (value INTEGER)')
engine.execute('INSERT INTO Fib VALUES (?)',[(str(x),) for x in fib(10)])
print(engine.execute("SELECT * FROM Fib").fetchall())
```
